## **Resample audio to 44.1 khz**

```python
sr_new = 44100
audio_new = librosa.resample(audio, orig_sr=sr, target_sr=sr_new)
# Pass original and target sampling rates as keyword arguments
```

## **Noice Reduction**

```python
import noisereduce as nr
audio_reduced=nr.reduce_noise(y=audio_new,sr=sr_new)
```

## **Normalize amplitde**

```python
import numpy as np
```

```python
audio_normalised=audio_reduced/np.max(np.abs(audio_reduced))
```
